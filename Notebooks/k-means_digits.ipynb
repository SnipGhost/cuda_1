```python
from time import time
import numpy as np
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
from sklearn.preprocessing import scale
```

```python
np.random.seed(42)

digits = load_digits()
data = scale(digits.data)
```

```python
idx = 100
img = digits.images[idx].astype(np.uint8)
trg = digits.target[idx]
```

```python
plt.figure(figsize=(2,2))
plt.axis("off")
plt.imshow(img, cmap='gray')
plt.title("Number: {}".format(str(trg)))
plt.show()
```

```python
kmean = KMeans(n_clusters=10, random_state=0).fit(data)
```

```python
proc_data = kmean.predict(data)
```

```python
idx = 4
```

```python
zero_cluster = digits.images[np.where(proc_data == idx)[0]]
zero_targets = digits.target[np.where(proc_data == idx)[0]]
```

```python
plt.figure()

for i in range(1, 17):
    plt.subplot(4, 4, i)
    
    img = zero_cluster[i].astype(np.uint8)
    trg = zero_targets[i]  
    
    plt.axis("off")
    plt.imshow(img, cmap='gray')
    plt.title("{}".format(str(trg)))
    
plt.tight_layout()
plt.show()
```
